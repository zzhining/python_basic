## 1 파이썬의 특징

파이썬은 배우고, 쓰기가 쉽습니다. 이것이 가장 큰 장점입니다. 그렇지만 프로그램 실행 속도가 컴파일 방식을 사용하는 언어에 비해서 느리기 때문에 불과 10여년 전까지도 프로그래머 사이에서 인기있는 언어가 아니었습니다. 그런데 인공지능(AI) 관련 기술이 발달하고, 대용량의 데이터 처리를 효율적으로 하려는 니즈가 많아지면서 파이썬에 대한 관심이 크게 늘었습니다. 파이썬이 이 분야에서 뛰어나기 때문입니다. 이러한 장점이 배우고 쓰기 쉽다는 기존의 장점과 함께 부각되면서 가장 주목받는 언어가 되었습니다.

### 조건문

컴퓨터에게 명령을 하기 위해서 조건을 정의해야하는 경우가 있습니다. ‘이런 경우에는 이렇게 해’라는 명령을 내리고 싶을 때, ‘이런 경우’라는 상황을 판단하는 로직을 정의할 때 사용하는 구문이 조건문입니다. 조건문은 `if`를 사용하여 정의합니다.

```
[if문]
- 지정한 조건에 따라 다르게 분기해 명령을 수행
- 조건의 만족 여부에 따라서 코드 수행 결과가 달라짐
```

`if`문은 참(True) 또는 거짓(False)을 판단하는 <조건문>과 함께 사용합니다.

<조건문>은 비교 연산자 또는 논리연산자를 사용하여 작성할 수 있습니다. 비교 연산자와 논리 연산자의 각 의미는 아래와 같습니다.

**[비교 연산자]**

| 비교 연산자 | 의미        | 활용 예   | 설명                    |
|-------------|-------------|-----------|-------------------------|
| ==          | 같다        | a == b    | a는 b와 같다            |
| !=          | 같지 않다   | a != b    | a는  b와  같지 않다     |
| &lt;        | 작다        | a &lt; b  | a는  b보다 작다         |
| &gt;        | 크다        | a &gt; b  | a는  b보다 크다         |
| &lt;=       | 작거나 같다 | a &lt;= b | a는  b보다  작거나 같다 |
| &gt;=       | 크거나 같다 | a &gt;= b | a는  b보다  크거나 같다 |


**[ 논리 연산자]**

| 논리 연산자 | 의미     | 활용 예 | 설명     |
|-------------|----------|---------|-----------------------------------------------------------------|
| and         | 논리곱   | A and B | A와  B가  모두  참이면  참이고    그  외에는  거짓              |
| or          | 논리합   | A or B  | A와  B  중  하나라도  참이면    참이고  둘  다  거짓이면   거짓 |
| not         | 논리부정 | not A   | A가  참이면  거짓이고    거짓이면  참                           |

 예를 들어서 특정 변수 `x`의 값이 90보다 크거나 같으면 ‘Pass’를 출력하는 코드를 작성해보겠습니다.

````python
# 변수 x의 값이 90보다 크거나 같으면 ‘Pass’를 출력하는 예
x = 95

if x >= 90:
    print("pass")
````

### if~else
<조건문>의 만족 여부에 따라 코드를 다르게 수행하려면 'if ~ else' 구조를 사용합니다. 예를 들어 변수 `x`의 값이 90이상인 경우에는 "Very good"을, 그렇지 않은 경우에는 "Bad"를 출력하고자 할 때에 아래와 같이 코드를 작성할 수 있습니다.

````python
x = 85
if x >= 90:
    print("Very good")
else:
    print("Bad")
````

### if ~ elif ~ else
여러 조건에 따른 분기를 하는 분기문을 작성할 경우에는 'if ~ elif ~ else' 구조를 사용합니다. `elif` 구문도 `if`문과 같이 조건문을 함께 작성합니다. 위의 코드에서 변수 `x`가 80이상, 90 미만인 경우에는 "Good"이라고 출력하는 로직을 추가해보겠습니다.


````python
x = 85
if x >= 90:
    print("Very good")
elif x >= 80 and x < 90:
    print("Good")
else:
    print("Bad")
````

### 중첩 조건문
하나의 조건문 안에 또 다른 조건을 판단하는 로직을 생성할 수 있습니다. 이 경우 각 구문의 들여쓰기를 사용하여 여러 개의 if문을 중첩하여 사용합니다. 위의 코드에서 변수 `x`가 90이상인 경우에 대해 `x`가 100인 경우에는 "Perfect"를 출력하고, 그렇지 않은 경우에는 "Very Good"을 출력하도록 코드를 수정해보겠습니다.

````python
x = 100
if x >= 90:
    if x==100 :
        print("Perfect")
    else:
        print("Very Good")
elif x >= 80 and x < 90:
    print("Good")
else:
    print("Bad")
````

튜플은 소괄호 ( ) 안에 여러 개의 아이템을 콤마(,)로 연결하여 만든 자료형입니다. 각각의 아이템은 리스트와 마찬가지로 순서를 가지고 있습니다. 하지만 튜플은 아이템을 선택하여 변경, 추가, 삭제가 불가능합니다. 아래 코드를 사용하여 Seoul, Tokyo, Beijing 세 개의 도시가 있는 `city_tuple` 튜플을 생성합니다.


````python
city_tuple = ('Seoul', 'Tokyo', 'Beijing')
````

튜플도 리스트처럼 아이템의 중복 생성을 허용합니다.


````python
city_tuple = ('Seoul', 'Tokyo', 'Beijing', 'Seoul', 'Tokyo')
print(city_tuple)
````

튜플도 인덱스를 사용하여 특정 위치의 아이템을 선택할 수 있습니다. 리스트와 동일한 방식으로 선택합니다. 1번 인덱스의 아이템을 선택해 보겠습니다.


````python
city_tuple = ('Seoul', 'Tokyo', 'Beijing')
city_tuple[1]
````

하지만 튜플은 편집이 불가능하기 때문에 특정 위치의 아이템을 변경할 수 없습니다. 아이템 값을 변경하려고 할 경우, 아래와 같이 에러가 발생합니다.


````python
# 아이템 변경 안됨
city_tuple[1] = 'Hanoi'
````

튜플은 아이템의 편집이 불가능하기 때문에 리스트에서 제공하는 `append()`, `remove()` 등의 기능을 제공하지 않습니다. 튜플에서 `append()` 또는 `remove()`를 호출하면 에러가 발생합니다.튜플의 아이템을 편집하기 위해서는 우선 리스트로 타입 변환을 진행하고 값을 추가하거나 제거한 후, 다시 튜플로 변환하는 과정이 필요합니다. 형 변환은 list(), tuple() 함수를 사용합니다.


````python
# 튜플을 리스트로 형 변환
city_list = list(city_tuple)

# 리스트의 값 편집:추가 및 삭제
city_list.append('London')
city_list.remove('Seoul')

# 리스트를 다시 튜플로 형변환
city_tuple = tuple(city_list)
print(city_tuple)
````

### 세트

**[Set의 집합 연산]**


세트는 집합의 성격을 가지고 있는 자료형이기 때문에 합집합, 교집합, 차집합 등 집합 연산을 위한 다양한 함수를 제공합니다.
| 연산      | 설명                                                                                |
|------------------------|-----------------------------------------------------------------------------------------------|
| difference()           | 차집합(두 세트 간의 차이를 반환)                                                              |

| symmetric_difference() | 두 세트의 합집합에서 교집합 부분을 제외한 부분을 반환(두 세트의 대칭 차이가 있는 세트를 반환) |
| union()                | 합집합                                                                                        |

### 딕셔너리
딕셔너리는 중괄호 { } 안에 여러 개의 키(key)-값(value) 쌍을 가지는 아이템을 콤마(,)로 연결하여 만든 자료형입니다. 각각의 아이템이 가지고 있는 키(key)를 활용해서 아이템의 값(value) 선택하거나 변경, 추가, 삭제가 가능합니다.

딕셔너리의 각 아이템은 키-값 쌍으로 이루어져 있습니다. 키-값 쌍은 콜론(:)을 사용하여 키:값 형태로 표시합니다. 아래 코드를 사용하여 name:Harry, age:27, height:190, weight:99.9 네 개의 아이템을 가지고 있는 `my_dict`딕셔너리를 생성합니다.


````python
my_dict = {
    'name': 'Harry',
    'age': 27,
    'height' : 190,
    'weight' : 99.9
}
print(my_dict)
````

딕셔너리는 각각의 아이템이 가지고 있는 키(key)를 활용해서 아이템의 값(value)을 선택할 수 있습니다. 또한 딕셔너리의 `keys()`를 호출하여 딕셔너리 객체가 가지고 있는 모든 키(key)값을 확인할 수 있습니다.

````python
# 딕셔너리 객체가 가지고 있는 모든 키(key)값을 확인
print(my_dict.keys())

# 키가 'age'인 아이템의 값을 확인
print(my_dict['age'])
````

딕셔너리의 키(key)를 사용하여 특정 아이템을 선택하고, 그 아이템의 값(value)을 변경할 수 있습니다.


````python
print(my_dict)
# 키가 'age'인 아이템의 값을 28로 할당
my_dict['age'] = 28
print(my_dict)
````

또한 딕셔너리의 `update()`를 사용하여 기존의 키-값 쌍을 업데이트하거나, 새로운 키-값 쌍을 생성할 수 있습니다.


````python
# 키가 'weight'인 아이템의 값을 100으로 할당
my_dict.update({'weight': 100})
print(my_dict)

# 키가 'address'인 아이템의 값을 Busan로 할당
my_dict.update({'address': 'Busan'})
print(my_dict)
````

딕셔너리의 아이템을 삭제하는 방법은 마지막 아이템을 삭제하는 `popitem()`, 특정 키 값을 지정하여 삭제하는 `pop()`, 딕셔너리 객체는 남겨두지만 안에 있는 아이템을 모두 삭제하는 `clear()`가 있습니다.


````python
#popitem(): 마지막 아이템 삭제
my_dict.popitem()
print(my_dict)
````

````python
# pop(): 특정 키('age') 값을 지정하여 아이템 삭제
my_dict.pop('age')
print(my_dict)
````

````python
# clear(): 딕셔너리 안에 있는 아이템을 모두 삭제
my_dict.clear()
print(my_dict)
````

### 반복문

특정 구문을 반복해서 수행해야 하는 경우가 있습니다. 예를 들어서 ‘주어진 1000개의 데이터의 값이 어떤 조건을 만족할 경우, 그 값을 삭제한다’라는 동작을 수행하는 프로그램을 반복문을 사용하지 않고 작성한다면 동일한 구문을 1000번 작성해야 합니다. 그러나 반복문을 사용하면 아주 간단하게 작성할 수 있습니다. 이번 절에서는 구문을 반복해서 수행하는 for문과 while문을 살펴봅니다.

반복문을 만드는 방법은 ‘for’를 사용하는 방법과 ‘while’을 사용하는 방법이 있습니다. 우선 for 구문의 살펴보겠습니다.
```
for [변수] in [반복자]:
(들여쓰기) 수행문
```
반복자(iterator)는 보유한 아이템을 순회할 수 있는 특징이 있습니다. 이해를 돕기 위해 반복자를 우리가 앞에서 배운 컨테이너 형을 떠올려서 생각해보세요. 컨테이너는 여러 개의 아이템을 담을 수 있는 자료형입니다. 컨테이너에 있는 여러 아이템을 차례대로 변수에 하나씩 담아서 수행문으로 넘겨서 문장을 수행합니다. 예를 들어서 1부터 5까지의 값이 담긴 리스트의 값을 하나씩 꺼내어 출력하는 코드를 작성하면 다음과 같습니다.

````python
for a in [0,1,2,3,4,5]:
    print(a)
````

반복자는 다양한 컨테이너 타입을 설정할 수 있지만 일반적으로 `리스트`와 `range()` 함수를 가장 많이 사용합니다.

````python
myFriends = ['James', 'Robert', 'Lisa', 'Mary']

for friend in myFriends:
    print(friend)
````

`range()`함수는 순차적으로 증가하는 값을 반복자로 사용하고자 할 때 사용합니다. `range()`함수의 매개변수는 다음과 같습니다.
```
range(start, stop, step)
- `start`는 범위의 시작 지점, `stop`은 범위의 끝 지점, `step`은 증감의 크기
- `step`이 1인 경우 생략 가능
- `step`이 1이고 `start`가 0인 경우 `start`역시 생략 가능
```

````python
# 0부터 10까지 1씩 증가하는 값을 반복자로 생성
print(range(0, 10, 1))


# 0부터 10까지 1씩 증가하는 값을 리스트로 생성
print(list(range(0,10,1)))

# 0부터 6까지 1씩 증가하는 값을 반복자로 생성하여 하나씩 값을 출력
for a in range(0,6):
    print(a)
````

`enumerate()`는 파이썬에서 반복(iterate) 가능한 객체(예: 리스트, 튜플, 문자열 등)를 열거할 때 인덱스와 요소를 동시에 추출할 수 있게 도와주는 내장 함수입니다. `enumerate()`를 사용하면 코드의 가독성을 높이고 반복문 내에서 인덱스를 수동으로 관리하는 번거로움을 줄일 수 있습니다.


```
enumerate(iterable, start=0)
- `iterable`: 반복 가능한 객체
- `start`: 인덱스의 시작 값을 지정, 기본값은 0
```

````python
x_list = ['a', 'b', 'c', 'd', 'e']
for m, n in enumerate(x_list):
    print(m, n)
````

for문의 반복자로 딕셔너리를 지정하면 딕셔너리의 키와 값을 확인할 수 있습니다.

````python
x_dict = {'a': 6, 'b': 2, 'c': 3, 'd': 7}

for i in x_dict:
    print(i)
````

딕셔너리의 키와 값 각각 받을 수 있도록 변수를 설정하면 키와 값을 구분하여 처리할 수 있습니다.

````python
for i, j in x_dict.items():
    print(i, j)
````

`for` 문도 `if`문과 마찬가지로 중첩된 형태로 사용할 수 있습니다. 예를 들어서 첫번째 리스트에 있는 값과 두번째 리스트에 담긴 값을 하나씩 꺼내어 모든 경우의 수를 조합하여 연산하는 경우에 중첩된 반복문의 작성이 필요합니다.

````python
x_list = ['x1', 'x2']
y_list = ['y1', 'y2']

print("xy")

for x in x_list:
    for y in y_list:
        print(x, y)
````

`while` 문은 조건에 따라 반복 여부를 결정한다는 점에 있어서 `for`문과 사용 방법에 차이가 있습니다. while 문은 조건문의 조건이 만족할 경우에 수행문을 반복해서 사용하고, 그렇지 않은 경우에는 while문을 빠져나오게 됩니다.
```
while [조건문]:
(들여쓰기) 수행문
```

````python
i= 0
sum = 0
print(i, sum)

while sum < 20:
    i = i + 1
    sum = sum + i
    print(i, sum)
````

`while` 문은 조건을 만족하는 경우, 무한 반복 수행하기 때문에 무한 반복 명령문을 작성할 수 있습니다. 의도적으로 주어진 코드 블록을 계속 반복하라고 명령을 내려야 할 때는 상관없지만, 실수로 무한 반복문을 작성하게 될 경우에는 프로세서에 부하가 걸리게되니 주의가 필요합니다.

````python
while True:
   print("while test")
````

`break`문은 반복문에서 빠져나가는 명령을 수행합니다. `for`문의 반복자를 모두 수행하지 않더라도 또한 `while`문의 조건이 참인 상태라고 하더라도 `break`문을 실행하게 되면 반복문은 더이상 수행되지 않고 종료됩니다.

````python
k = 0

while True:
    k = k + 1
    if k > 3:
        break
    print(k)
````

`continue`는 해당 차례의 남은 작업을 하지 않고 다음 반복으로 넘어가는 구문입니다. `break`와 다르게 반복문이 종료되는 것이 아니고, 그 다음 반복 조건에 해당하는 구문을 수행한다는 점에 있어서 차이가 있습니다.

````python
for k in range(5):
    if k==2:
        continue
    print(k)
````

#### 컴프리헨션 (comprehension)

컴프리헨션(Comprehension)은 파이썬에서 리스트, 집합, 딕셔너리 등의 컨테이너 타입의 자료형을 생성하는 간결하고 효율적인 방법입니다. 컴프리헨션은 기존의 컨테이너 타입의 자료형을 기반으로 새로운 컨테이너형 자료를 만들 때 유용하며, 코드의 가독성을 높이고 작성 시간을 줄여줍니다. 파이썬에서는 주로 리스트 컴프리헨션(List Comprehension), 딕셔너리 컴프리헨션(Dictionary Comprehension), 집합 컴프리헨션(Set Comprehension)이 사용됩니다.

**리스트 컴프리헨션 (List Comprehension)(())
리스트 컴프리헨션은 기존 리스트를 변형하여 새로운 리스트를 만들기 위한 구문입니다.
```
`[<반복 실행문> for <반복 변수> in <반복 범위>]`
```

````python
numbers = [1,2,3,4,5]
square = [i **2 for i in numbers]
print(square)
````

조건문을 포함한 리스트 컴프리헨션의 문법은 아래와 같습니다.



`[<반복 실행문> for <반복 변수> in <반복 범위> if <조건문>]`


````python
square = [i ** 2 for i in numbers if i >= 3]
print(square)
````

else가 포함된 경우의 리스트 컴프리헨션의 문법은 아래와 같습니다.


`[<반복 실행문> if <조건문> else <else문> for <반복 변수> in <반복 범위> ]`

````python
test = [x if x < 5 else x * 3 for x in range(10)]
print(test)
````

### 파일 I/O

**파일 열기**

파이썬에서 파일을 열기 위해서는 내장 함수 `open()`을 사용합니다. 예를 들어, 다음과 같이 파일을 열 수 있습니다.

```
f = open('file_name', 'mode')
```
여기서 `mode`는 파일을 열 때 사용할 모드를 지정하며, 다음과 같은 모드를 사용할 수 있습니다. 기본적으로 파일을 열 때 `r` 모드를 사용하여 읽기 모드로 열립니다. `w` 모드는 쓰기 모드로 파일을 열며, 같은 이름의 파일이 있으면 기존 내용은 모두 삭제됩니다. `x` 모드는 쓰기 모드로 파일을 열지만, 같은 이름의 파일이 있을 경우 오류가 발생합니다. `a` 모드는 추가 모드로 파일을 열며, 같은 이름의 파일이 없으면 쓰기 모드와 기능이 같습니다. `b` 모드는 바이너리 파일 모드로 파일을 열고, `t` 모드는 텍스트 파일 모드로 파일을 엽니다. `mode`는 혼합해서 사용할 수도 있습니다.

**파일 쓰기**
파일에 데이터를 쓰기 위해서는 파일을 쓰기 모드로 열어야 합니다. 파일을 열고 지정한 내용을 쓴 후에는 파일을 닫아야 합니다. 예를 들어, 다음 코드는 파일을 열고 문자열을 쓴 후 파일을 닫는 과정을 보여줍니다.


````python
f = open('myFile.txt', 'w')         # 'myFile.txt' 파일을 쓰기 모드로 열기
f.write('This is my first file.')   # 연 파일에 문자열 쓰기
f.close()                           # 파일 닫기
````

**파일 읽기**

파일을 읽으려면 파일을 읽기 모드로 열어야 합니다. 파일의 내용을 읽고 마지막으로 파일을 닫습니다. 예를 들어, 다음 코드는 파일을 읽고 내용을 출력하는 과정을 보여줍니다.

````python
f = open('myFile.txt', 'r')     # 'myFile.txt' 파일을 읽기 모드로 열기
file_text = f.read()            # 파일 내용 읽은 후 변수에 저장
f.close()                       # 파일 닫기

print(file_text)                # 변수에 저장된 내용 출력
````


파일을 읽을 때 `read()` 메서드에 `count` 매개변수를 지정하면 시작 위치부터 해당 바이트까지 읽을 수 있습니다. 파일을 열 때 인코딩을 지정할 수 있습니다. 예를 들어, 다음 코드는 `cp949` 인코딩을 사용하여 파일을 작성하고 읽는 방법을 보여줍니다:

````python
f = open("myFile_ko.txt", "w", encoding="cp949")  # cp949 인코딩으로 파일 쓰기 모드로 열기
f.write("텍스트 파일을 새로 만들고 텍스트를 씁니다.\n")  # 파일에 텍스트 쓰기
f.close()  # 파일 닫기

f = open('myFile_ko.txt', 'r', encoding="cp949")  # cp949 인코딩으로 파일 읽기 모드로 열기
file_text_ko = f.read()  # 파일 내용 읽기
f.close()  # 파일 닫기

print(file_text_ko)  # 파일 내용 출력
````


파일에 문자열을 한 줄씩 쓰기 위해서는 다음과 같이 작성할 수 있습니다.

````python
f = open('two_times_table.txt', 'w')  # 파일을 쓰기 모드로 열기
for num in range(1, 6):  # num이 1에서 5까지 반복
    format_string = "2 x {0} = {1}\n".format(num, 2*num)  # 저장할 문자열 생성
    f.write(format_string)  # 파일에 문자열 저장
f.close()  # 파일 닫기
````

**파일에서 문자열 한 줄씩 읽기**

파일로부터 문자열을 한 줄씩 읽기 위해서는 `readline()` 메서드를 사용할 수 있습니다. 마지막 줄을 읽은 후 다시 `readline()`을 실행하면 빈 문자열을 반환합니다. 예를 들어, 다음 코드는 파일에서 두 줄을 읽는 과정을 보여줍니다:


````python
f = open("two_times_table.txt")  # 파일을 읽기 모드로 열기
line1 = f.readline()  # 한 줄씩 문자열을 읽기
line2 = f.readline()
f.close()  # 파일 닫기

print(line1, end="")  # 한 줄씩 문자열 출력(줄 바꿈 안 함)
print(line2, end="")
````

파일 전체에서 한 줄씩 읽어 오기 위해서는 다음과 같이 작성할 수 있습니다.

````python
f = open("two_times_table.txt")  # 파일을 읽기 모드로 열기
line = f.readline()  # 문자열 한 줄 읽기
while line:  # line이 공백인지 검사해서 반복 여부 결정
    print(line, end="")  # 문자열 한 줄 출력(줄 바꿈 안 함)
    line = f.readline()  # 문자열 한 줄 읽기
f.close()  # 파일 닫기
````

파일 전체의 모든 줄을 읽어서 한 줄씩을 요소로 갖는 리스트 타입으로 반환하려면 `readlines()` 메서드를 사용합니다.

````python
f = open("two_times_table.txt")  # 파일을 읽기 모드로 열기
lines = f.readlines()  # 파일 전체 읽기(리스트로 반환)
f.close()  # 파일 닫기

print(lines)  # 리스트 변수 내용 출력
````

리스트에 할당된 문자열을 for 문을 이용해 항목을 하나씩 처리할 수 있습니다.

````python
f = open("two_times_table.txt")  # 파일을 읽기 모드로 열기
lines = f.readlines()  # 파일 전체 읽기(리스트로 반환)
f.close()  # 파일 닫기

for line in lines:  # 리스트를 반복 범위로 지정
    print(line, end="")  # 리스트 항목을 출력(줄 바꿈 안 함)
````

for 문의 반복 범위에 `lines` 변수 대신 바로 `f.readlines()`를 사용할 수도 있습니다.

````python
f = open("two_times_table.txt")  # 파일을 읽기 모드로 열기
for line in f.readlines():  # 파일 전체를 읽고, 리스트 항목을 line에 할당
    print(line, end="")  # 리스트 항목을 출력(줄 바꿈 안 함)
f.close()  # 파일 닫기
````

#### with 문

파일 작업을 할 때 `with` 문을 사용하면 파일을 열고 닫는 과정을 자동으로 처리할 수 있어 코드가 더 깔끔해집니다. 예를 들어, 다음 코드는 `with` 문을 이용해 파일을 열고 문자열을 쓰는 과정을 보여줍니다.

````python
with open('myTextFile2.txt', 'w') as f:  # 파일 열기
    f.write('File read/write test2: line1\n')  # 파일 쓰기
    f.write('File read/write test2: line2\n')
    f.write('File read/write test2: line3\n')
````

with 문을 이용해 생성한 파일을 읽는 예제는 다음과 같습니다.

````python
with open('myTextFile2.txt') as f:  # 파일 열기
    file_string = f.read()  # 파일 읽기
    print(file_string)
````

with 문을 반복문과 함께 이용해 파일의 문자열을 한 줄씩 읽는 예제는 다음과 같습니다.

````python
with open('myTextFile3.txt', 'w') as f:  # 파일을 쓰기 모드로 열기
    for num in range(1, 6):  # for 문에서 num이 1에서 5까지 반복
        format_string = "3 x {0} = {1}\n".format(num, 3*num)  # 문자열 생성
        f.write(format_string)  # 문자열을 파일에 쓰기
````

with 문과 for 문을 이용해 파일의 문자열을 한 줄씩 읽는 예제는 다음과 같습니다.

````python
with open('myTextFile3.txt', 'r') as f:  # 파일을 읽기 모드로 열기
    for line in f:  # 파일 전체를 읽고 리스트 항목을 line에 할당
        print(line, end="")  # line에 할당된 문자열 출력(줄 바꿈 안 함)
````

````python
# -*- coding: utf-8 -*-
"""13_object.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/github/zzhining/python_basic/blob/master/13_object.ipynb

# 객체지향프로그래밍(OOP)

- 객체를 만들고 이용할 수 있는 기능을 제공하는 프로그래밍 언어

# 객체
 - *속성*과 *행위*로 구성
 - **속성**: 특징, 상태 → 변수로 구현
 - **행위**: 할 수 있는 일(행동, 동작, 기능) → 함수로 구현
 - 객체는 변수와 함수의 묶음

# 객체 만드는 방법
- 객체를 만들려면 먼저 **클래스**를 선언해야 함
- 클래스: 객체의 공통된 속성과 행위를 변수와 함수로 정의한 것
- **클래스**는 객체를 만들기 위한 기본 틀, **객체**는 기본 틀을 바탕으로 만들어진 결과
- 객체는 클래스에서 생성하므로 객체를 클래스의 인스턴스 (Instance)라고 함
----
⭐클래스와 객체의 관계⭐

 1. 클래스 선언(붕어빵 틀)
     - 클래스: 객체를 만들기 위한 기본 틀
 2. 객체 생성(붕어빵)
     - 인스턴스(instance)

![class](https://blog.kakaocdn.net/dn/chH1CQ/btrp55RgLk4/B5LFxkHKOjWFwkOOvBuKt0/img.png)

클래스 선언을 위한 기본 구조
```
class 클래스명():
   [변수1]  # 클래스 변수
   [변수2]
   ...
   def 함수1(self[, 인자1, 인자2, ··· , 인자n]): # 클래스 함수
       <코드 블록>
       ...
   def 함수2(self[, 인자1, 인자2, ··· , 인자n]):
       <코드 블록>
       ...
```

## [예제]: 자전거 클래스

- 클래스 이름: Bicycle()
- 속성: 바퀴크기(wheelSize), 색상(color)
- 동작: 이동(move), 회전(turn), 정지(stop)
"""

# 자전거 클래스 선언
class Bicycle(): # 클래스 선언
    pass

"""객체 초기화
- 초기화 함수 `__init__()`를 구현하면 객체를 생성하는 것과 동시에 속성값을 지정할 수 있음
- `__init__()` 함수는 클래스의 인스턴스가 생성될 때 (즉, 객체가 생성될 때) 자동으로 실행됨

속성 및 동작 추가
"""

class Bicycle():

    count = 0

    def __init__(self, wheelSize, color):
        self.wheelSize = wheelSize
        self.color = color
        Bicycle.count = Bicycle.count + 1
        print("[init] count = {0}".format(Bicycle.count))

    def move(self, speed):
        print("[move] speed = {0}".format(speed))

    def turn(self, direction):
         print("[turn] direction = {0}".format(direction))

    def stop(self):
         print("[stop] wheelSize = {0}, color={1}".format(self.wheelSize, self.color))

"""객체 선언"""

myBicycle = Bicycle(50, 'Blue')

myBicycle2 = Bicycle(20, 'Red')

"""메서드 호출"""

myBicycle.move(20)

# myBicycle.wheelSize = 30
# myBicycle.color = 'Red'
myBicycle.stop()

"""# 클래스에서  사용하는 변수
- 위치에 따라 **클래스 변수 (class variable)**와 **인스턴스 변수 (instance variable)**로 구분
- **클래스 변수**: 클래스 내에 있지만 함수 밖에서 `'변수명 = 데이터'` 형식으로 정의한 변수
    - 클래스에서 생성한 모든 객체가 공통으로 사용 가능
    - '클래스명.변수명' 형식으로 접근
- **인스턴스 변수**: 클래스 내의 함수 안에서 `'self.변수명= 데이터'` 형식으로 정의한 변수
    - 클래스 내의 모든 함수에서 'self.변수명'으로 접근
    - 각 인스턴스(객체)에서 개별적으로 관리하며, 객체를 생성한 후에 '객체명.변수명' 형식으로 접근



"""

# 클래스 변수와 인스턴스 변수를 사용한 자동차 클래스
class Car():
    instance_count = 0 # 클래스 변수 생성 및 초기화

    def __init__(self, size, color):
        self.size = size    # 인스턴스 변수 생성 및 초기화
        self.color = color  # 인스턴스 변수 생성 및 초기화
        Car.instance_count = Car.instance_count + 1 # 클래스 변수 이용
        print("자동차 객체의 수: {0}".format(Car.instance_count))

    def move(self):
        print("자동차({0} & {1})가 움직입니다.".format(self.size, self.color))

car1 = Car('small', 'white')
car2 = Car('big', 'black')

print("Car 클래스의 총 인스턴스 개수:{}".format(Car.instance_count))
print("Car 클래스의 총 인스턴스 개수:{}".format(car1.instance_count))
print("Car 클래스의 총 인스턴스 개수:{}".format(car2.instance_count))

car1.move()
car2.move()

# 이름이 같은 클래스 변수와 인스턴스 변수가 있는 클래스를 정의한 경우
class Car2():
    count = 0; # 클래스 변수 생성 및 초기화

    def __init__(self, size, num):
        self.size = size    # 인스턴스 변수 생성 및 초기화
        self.count = num  # 인스턴스 변수 생성 및 초기화
        Car2.count = Car2.count + 1 # 클래스 변수 이용
        print("자동차 객체의 수: Car2.count = {0}".format(Car2.count))
        print("인스턴스 변수 초기화: self.count = {0}".format(self.count))

    def move(self):
        print("자동차({0} & {1})가 움직입니다.".format(self.size, self.count))

car1 = Car2("big", 20)
car1 = Car2("small", 30)

"""# 클래스에서 사용하는 함수

1. **인스턴스 메서드(instance method)**
    - 각 객체에서 개별적으로 동작하는 함수를 만들고자 할 때 사용
    - 함수를 정의할 때 첫 인자로 self가 필요
    - 인스턴스 메서드 안에서는 `self.함수명()` 형식으로 클래스 내의 다른 함수 호출


2. **정적 메서드(static method)**
    - 클래스와 관련이 있어서 클래스 안에 두기는 하지만,클래스나 클래스의 인스턴스와는 무관하게 독립적으로 동작하는 함수
    - self를 사용하지 않음
    - 정적메서드 안에서는 클래스나 클래스 변수에 접근할 수 없음
    - 데코레이터: @staticmethod


3. **클래스 메서드(class method)**
    - 클래스 변수를 사용하기 위한 함수
    - 함수를 정의할 때 첫 번째 인자로 클래스를 넘겨받는 cls가 필요
    - 함수 앞에 데코레이터인 @classmethod를 지정

    ---------------

## 인스턴스 메서드
```
객체명 = 클래스명()
객체명.메서드명([인자1, 인자2, ··· , 인자n])
```

인스턴스 메서드를 사용한 자동차 클래스
"""

# Car 클래스 선언
class Car():
    instance_count = 0 # 클래스 변수 생성 및 초기화

    # 초기화 함수(인스턴스 메서드)
    def __init__(self, size, color):
        self.size = size    # 인스턴스 변수 생성 및 초기화
        self.color = color  # 인스턴스 변수 생성 및 초기화
        Car.instance_count = Car.instance_count + 1 # 클래스 변수 이용
        print("자동차 객체의 수: {0}".format(Car.instance_count))

    # 인스턴스 메서드
    def move(self, speed):
        self.speed = speed  # 인스턴스 변수 생성
        print("자동차({0} & {1})가 ".format(self.size, self.color), end='')
        print("시속 {0}킬로미터로 전진".format(self.speed))

    # 인스턴스 메서드
    def auto_cruise(self):
        print("자율 주행 모드")
        self.move(self.speed) # move() 함수의 인자로 인스턴스 변수를 입력

"""객체를 생성하고 인스턴스 메서드를 사용하는 예"""

car1 = Car("small", "red") # 객체 생성 (car1)
car2 = Car("big", "green") # 객체 생성 (car2)

car1.move(80) #객체(car1)의 move() 메서드 호출
car2.move(100) #객체(car2)의 move() 메서드 호출

car1.auto_cruise() #객체(car1)의 auto_cruise() 메서드 호출
car2.auto_cruise() #객체(car2)의 auto_cruise() 메서드 호출

"""## 정적 메서드
- 클래스와 관련이 있어서 클래스 안에 두기는 하지만 클래스나 클래스의  인스턴스(객체)와는 무관하게 독립적으로 동작하는 함수를 만들고 싶을 때 이용하는 함수
- 함수를 정의할 때 인자로 self를 사용하지 않으며 정적 메서드 안에서는 클래스나 클래스 변수에  접근할 수 없음
- 함수 앞에 데코레이터(Decorator)인 @staticmethod를 선언해 정적 메서드임을  표시

정적 메서드의 구조

```
 class 클래스명():
       @staticmethod
       def 함수명([인자1, 인자2, ··· , 인자n]):
           <코드 블록>
```

정적 메서드 호출
```
클래스명.메서드명([인자1, 인자2, ··· , 인자n]):
```

정적 메서드를 사용한 예
"""

# Car 클래스 선언
class Car():

    # def __init__(self, size, color): => 앞의 코드 활용
    # def move(self, speed): => 앞의 코드 활용
    # def auto_cruise(self): => 앞의 코드 활용

    # 정적 메서드
    @staticmethod
    def check_type(model_code):
        if(model_code >= 20):
            print("이 자동차는 전기차입니다.")
        elif(10 <= model_code < 20):
            print("이 자동차는 가솔린차입니다.")
        else:
            print("이 자동차는 디젤차입니다.")

Car.check_type(25)
Car.check_type(2)

"""##클래스 메서드
- 클래스 변수를 사용하기 위한 함수
- 함수를 정의할 때 첫 번째 인자로 클래스를 넘겨받는 cls가 필요
- 함수 앞에 데코레이터인 @classmethod를 지정

클래스 메서드의 구조
```
 class 클래스명():
      @classmethod
      def 함수명(cls[, 인자1, 인자2, ··· , 인자n]):
          <코드 블록>
```

클래스 메서드를 호출하는 방법
```
클래스명.메서드명([인자1, 인자2, ··· , 인자n]):
```

클래스 메서드를 사용하는 예
"""

# Car 클래스 선언
class Car():
    instance_count = 0 # 클래스 변수

    # 초기화 함수(인스턴스 메서드)
    def __init__(self, size, color):
        self.size = size    # 인스턴스 변수
        self.color = color  # 인스턴스 변수
        Car.instance_count = Car.instance_count + 1

    # 클래스 메서드
    @classmethod
    def count_instance(cls):
        print("자동차 객체의 개수: {0}".format(cls.instance_count))

"""클래스 메서드를 사용하는 예"""

Car.count_instance()           # 객체 생성 전에 클래스 메서드 호출
car1 = Car("small", "red")     # 첫 번째 객체 생성
Car.count_instance()           # 클래스 메서드 호출
car2 = Car("big", "green")     # 두 번째 객체 생성
Car.count_instance()           # 클래스 메서드 호출

"""
## [예제]자동차 클래스(full-code)"""

class Car():
    instance_count = 0 # 클래스 변수 생성 및 초기화

    #초기화 함수(인스턴스 메서드)
    def __init__(self, size, color):
        self.size = size #인스턴스 변수 생성 및 초기화
        self.color = color # 인스턴스 변수 생성 및 초기화
        Car.instance_count = Car.instance_count + 1 # 클래스 변수
        print("자동차 객체 수: {0}".format(Car.instance_count))

    #인스턴스 메서드
    def move(self, speed) :
        self.speed = speed # 인스턴스 변수 생성
        print("자동차 {0} & {1}가".format(self.size, self.color), end='')
        print("시속 {0}km로 전진".format(self.speed))

    #인스턴스 메서드
    def auto_cruise(self):
        print("자율주행모드")
        self.move(self.speed) # 함수의 인자로 인스턴스 변수를 입력

    #정적메서드
    @staticmethod
    def check_type(model_code):
        if(model_code > 20):
            print("이 자동차는 전기차입니다")
        elif(10 <= model_code < 20) :
            print("이 자동차는 가솔린차입니다")
        else:
            print("이 자동차는 디젤차입니다")

    #클래스메서드
    @classmethod
    def count_instance(cls):
        print("자동차 객체의 개수: {0}".format(cls.instance_count))

car1 = Car("small", "red") # 객체생성(car1)
car2 = Car("big", "green") # 객체생성(car2)

print("--")
car1.move(80) #객체(car1)의 move() 메서드 호출
car2.move(100)#객체(Car2)의 move() 메서드 호출
print("--")
car1.auto_cruise() #객체(car1)의 auto_cruise 메서드 호출
car2.auto_cruise() #객체(car2)의 auto_cruise 메서드 호출
print("--")

Car.check_type(25)
Car.check_type(2)

Car.count_instance()

"""# 객체와 클래스를 사용하는 이유
- 코드 작성과 관리가 편하기 때문
- 규모가 큰 프로그램을 만들 때 클래스와 객체를 많이 이용
- 유사한 객체가 많은 프로그램을 만들 때도 주로 클래스와 객체를 이용해 코드를 작성

## [예제] 컴퓨터 게임의 로봇
- 로봇의 속성과 동작
    - 로봇의  속성: 이름, 위치
    - 로봇의  동작: 한 칸 이동

### 클래스와 객체를 사용하지 않는 코드
"""

robot_name = 'R1'   # 로봇 이름
robot_pos = 0     # 로봇의 초기 위치

def robot_move():
    global robot_pos
    robot_pos = robot_pos + 1
    print("{0} position: {1}".format(robot_name, robot_pos))

robot_move()

"""로봇을 추가해 두 대의 로봇을 구현"""

robot1_name = 'R1'   # 로봇 이름
robot1_pos = 0        # 로봇의 초기 위치

def robot1_move():
    global robot1_pos
    robot1_pos = robot1_pos + 1
    print("{0} position: {1}".format(robot1_name, robot1_pos))

#####################################################################
robot2_name = 'R2'    # 로봇 이름
robot2_pos = 10       # 로봇의 초기 위치

def robot2_move():
    global robot2_pos
    robot2_pos = robot2_pos + 1
    print("{0} position: {1}".format(robot2_name, robot2_pos))

robot1_move()
robot2_move()

"""### 클래스와 객체를 사용하는 코드"""

class Robot():
    def __init__(self, name, pos):
        self.name = name        # 로봇 객체의 이름
        self.pos = pos          # 로봇 객체의 위치

    def move(self):
        self.pos = self.pos + 1
        print("{0} position: {1}".format(self.name, self.pos))

robot1 = Robot('R1', 0)
robot2 = Robot('R2', 10)

robot1.move()
robot2.move()

myRobot3 = Robot('R3', 30)
myRobot4 = Robot('R4', 40)

myRobot3.move()
myRobot4.move()

"""# 상속

- 상속: 이미 만들어진 클래스의 변수와 함수를 그대로 이어받고 새로운 내용만 추가해서 클래스를 선언
- 상속관계에 있는 두 클래스는 자식이 부모의 유전적 형질을 이어받는 관계와 유사하기 때문에  흔히 부모 자식과의 관계로 표현

## 부모 클래스와 자식 클래스
- 부모 클래스: 상위  클래스 혹은 슈퍼클래스
- 자식 클래스: 하위 클래스 혹은 서브 클래스
- 자식 클래스가 부모 클래스로부터 상속을 받으면 자식 클래스는 부모 클래스의 속성(변수)과  행위(함수)를 그대로 이용 가능
- 상속 후에는 자식 클래스만 갖는 속성과 행위를 추가할 수  있음

## 부모 클래스와 자식 클래스의 관계<br>
<img src="https://blog.kakaocdn.net/dn/cycaBw/btqEkNO0T2Y/PQFOWxKG4ilXB7dmBVXABK/img.png" width="500" height="300">

부모 클래스를 상속받는 자식 클래스를 선언하는 형식
```
 class 자식 클래스 이름(부모 클래스 이름):
       <코드 블록>
```

##[예제] Bicycle을 상속하는 FoldingBicycle 클래스
"""

class FoldingBicycle(Bicycle):
    def __init__(self, wheel_size, color, state):
#         Bicycle.__init__(self, wheel_size, color)
        super().__init__(wheel_size, color)
        self.state = state

    def fold(self):
        self.state='folding'
        print("자전거: 접기, state={0}".format(self.state))

    def unfold(self):
        self.state='unfolding'
        print("자전거: 펴기, state={0}".format(self.state))

"""FoldingBicycle 클래스의  인스턴스를 생성한 후 메서드 호출"""

folding_bicycle = FoldingBicycle(27, 'white', 'unfolding') # 객체 생성
folding_bicycle.move(20)        # 부모 클래스의 함수(메서드) 호출
folding_bicycle.fold()          # 자식 클래스에서 정의한 함수 호출
folding_bicycle.unfold()

````
