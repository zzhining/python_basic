# 반복문
- 작업을 반복적으로 수행하는 구문
- `for` 문과 `while` 문

# for문

```python
for a in [0,1,2,3,4,5]:
    print(a)
```

## 반복 범위 지정
`리스트`와 `range()` 함수를 이용


```python
myFriends = ['James', 'Robert', 'Lisa', 'Mary']

for friend in myFriends:
    print(friend)
```

### range(start, stop, step)
- `start`는 범위의 시작 지점, `stop`은 범위의 끝 지점, `step`은 증감의 크기
- `step`이 1인 경우 생략 가능
- `step`이 1이고 `start`가 0인 경우 `start`역시 생략 가능



```python
print(range(0, 10, 1))
```

```python
print(list(range(0,10,1)))
```

```python
for a in range(0,6):
    print(a)
```

```python
x_list = ['a', 'b', 'c', 'd', 'e']
for m, n in enumerate(x_list):
    print(m, n)
```

딕셔너리의 키/값 출력


```python
x_dict = {'a': 6, 'b': 2, 'c': 3, 'd': 7}

for i in x_dict:
    print(i)
```

```python
for i, j in x_dict.items():
    print(i, j)
```

# 중첩 for 문

```python
x_list = ['x1', 'x2']
y_list = ['y1', 'y2']

print("xy")

for x in x_list:
    for y in y_list:
        print(x, y)
```

# while 문
- 조건에 따라 반복 여부를 결정


```python
i= 0
sum = 0
print(i, sum)

while sum < 20:
    i = i + 1
    sum = sum + i
    print(i, sum)
```

## 무한 반복 while 문
- 코드 블록을 무조건 계속 반복하라고 명령을 내려야 할 때

```python
while True:
   print("while test")
```

# break
- 반복문에서 빠져나옴

```python
k = 0

while True:
    k = k + 1
    if k > 3:
        break
    print(k)
```

## continue
- 해당 차례의 남은 작업을 하지 않고 다음 반복으로 넘어감

```python
for k in range(5):
    if k==2:
        continue
    print(k)
```

# 컴프리헨션 (comprehension)
- 리스트 컴프리헨션
- 세트 컴프리헨션
- 딕셔너리 컴프리헨션


## 리스트 컴프리헨션의 기본 구조

`[<반복 실행문> for <반복 변수> in <반복 범위>]`

```python
numbers = [1,2,3,4,5]
square = [i **2 for i in numbers]
print(square)
```

## 조건문을 포함한 리스트 컴프리헨션
`[<반복 실행문> for <반복 변수> in <반복 범위> if <조건문>]`

```python
square = [i ** 2 for i in numbers if i >= 3]
```
